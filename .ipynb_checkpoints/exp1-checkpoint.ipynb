```julia
function run_bm(num_obs, num_p, is_replay)
    N = num_obs
    total = @elapsed sample(big_hmm, SMC(num_p, is_replay))
    df = readtable("temp.csv", separator = ',', header = false)
    fork = sum([df[1,i] for i = 1:(length(df[1,:])-1)])
    fork / total
end
# ratio = run_bm(100, 20, true)
# println("$(ratio*100)%")
;
```

Define a helper function to collection counts from `Profile`.

```julia
function gen_counts()
    f = open("/Users/kai/temp/prof.txt","w")
    Profile.print(f, format=:flat)
    close(f)

    f = open("/Users/kai/temp/profdepth0.txt","w")
    Profile.print(f, maxdepth=0)
    close(f)
    
    fork_count = 0
    total_count = 0

    f = open("/Users/kai/temp/prof.txt")
    for ln in eachline(f)
        if contains(ln, "fork")
            toidx = search(ln, r"\w\s")[1] # find the first space after a character
            fork_count += parse(Int, ln[1:toidx]) 
        end
    end
    close(f)

    f = open("/Users/kai/temp/profdepth0.txt")
    for ln in eachline(f)
        if ln[2] != 'C' 
            toidx = search(ln, r"\w\s")[1]
            total_count += parse(Int, ln[1:toidx]) 
        end
    end
    close(f)

    return fork_count, total_count
end
;
```

Profile steps
1. Change the copying method in the source code
2. Run the block below
3. Copy the counting output to the next block

```julia
function run_exp(batch_size, p_num, obs_num)
    global N
    N = obs_num
    forks = []
    totals = []

    for _ = 1:batch_size
        Profile.init(delay=0.01)
        Profile.clear()
        @profile sample(big_hmm, SMC(p_num, true))
        fork_count, total_count = gen_counts()
        push!(forks, fork_count)
        push!(totals, total_count)
    end

    return (forks, totals)
end
;
```

```julia
ratios = []
for obs_num = 300:300:3000
    forks, totals = run_exp(2, 20, obs_num)
    ratio = sum(forks) / sum(totals)
    println(ratio)
    push!(ratios, ratio)
end
ratios
```

```julia
forks_process = []
for i = 1:10
    df = readtable("./exp1/probc/fork$i.csv", separator = ',', header = false)
    push!(forks_process, sum([df[1,i] for i = 1:(length(df[1,:])-1)]))
end
# totals_process = [17300,18811,18104,17864,18827,18051,18474,18355,18170,15699]
totals_process = []
df = readtable("./exp1/probc/total.txt", header = true)
for e in df[:,1]
    push!(totals_process, e)
end

# 200
forks_coroutine = [1,3,3,1,4,1,2,2,5,2]
totals_coroutine = [46,48,41,40,39,36,42,38,39,39]

# 300
forks_coroutine_2 = [4,1,2,1,3,3,2,3,4,4]
totals_coroutine_2 = [73,60,59,59,67,65,62,71,59,60]

# 400
forks_coroutine_3 = [6,7,6,5,3,1,4,7,5,5]
totals_coroutine_3 = [86,93,83,96,82,85,78,78,85,77]

# 200
forks_replaying = [60,52,44,46,42,50,54,50,41,28]
totals_replaying = [488,433,439,456,427,426,457,443,428,438]

# 300
forks_replaying_2 = [116,116,104,106,96,149,94,94,100,82]
totals_replaying_2 = [919,958,880,957,921,977,986,925,859,906]

# 400
forks_replaying_3 = [196,162,198,168,206,190,166,166,192,176]
totals_replaying_3 = [1636,1662,1642,1581,1667,1615,1549,1623,1621,1609]
;
```

Creating stats for plotting.

```julia
fracs_process = [Float64(forks_process[i] / totals_process[i]) for i = 1:1]
fracs_process_2 = [Float64(forks_process[i] / totals_process[i]) for i = 1:1]
fracs_process_3 = [Float64(forks_process[i] / totals_process[i]) for i = 1:1]
fracs_replaying = [Float64(forks_replaying[i] / totals_replaying[i]) for i = 1:10]
fracs_replaying_2 = [Float64(forks_replaying_2[i] / totals_replaying_2[i]) for i = 1:10]
fracs_replaying_3 = [Float64(forks_replaying_3[i] / totals_replaying_3[i]) for i = 1:10]
fracs_coroutine = [Float64(forks_coroutine[i] / totals_coroutine[i]) for i = 1:10]
fracs_coroutine_2 = [Float64(forks_coroutine_2[i] / totals_coroutine_2[i]) for i = 1:10]
fracs_coroutine_3 = [Float64(forks_coroutine_3[i] / totals_coroutine_3[i]) for i = 1:10]

means_process = mean(fracs_process)
means_process_2 = mean(fracs_process_2)
means_process_3 = mean(fracs_process_3)
means_replaying = mean(fracs_replaying)
means_replaying_2 = mean(fracs_replaying_2)
means_replaying_3 = mean(fracs_replaying_3)
means_coruntine = mean(fracs_coroutine)
means_coruntine_2 = mean(fracs_coroutine_2)
means_coruntine_3 = mean(fracs_coroutine_3)

stds_process = std(fracs_process)
stds_process_2 = std(fracs_process_2)
stds_process_3 = std(fracs_process_3)
stds_replaying = std(fracs_replaying)
stds_replaying_2 = std(fracs_replaying_2)
stds_replaying_3 = std(fracs_replaying_3)
stds_coruntine = std(fracs_coroutine)
stds_coruntine_2 = std(fracs_coroutine_2)
stds_coruntine_3 = std(fracs_coroutine_3)

mins_process = means_process - (1.96 * stds_process / sqrt(10))
mins_process_2 = means_process_2 - (1.96 * stds_process_2 / sqrt(10))
mins_process_3 = means_process_3 - (1.96 * stds_process_3 / sqrt(10))
mins_replaying = means_replaying - (1.96 * stds_replaying / sqrt(10))
mins_replaying_2 = means_replaying_2 - (1.96 * stds_replaying_2 / sqrt(10))
mins_replaying_3 = means_replaying_3 - (1.96 * stds_replaying_3 / sqrt(10))
mins_coruntine = means_coruntine - (1.96 * stds_coruntine / sqrt(10))
mins_coruntine_2 = means_coruntine_2 - (1.96 * stds_coruntine_2 / sqrt(10))
mins_coruntine_3 = means_coruntine_3 - (1.96 * stds_coruntine_3 / sqrt(10))

maxs_process = means_process + (1.96 * stds_process / sqrt(10))
maxs_process_2 = means_process_2 + (1.96 * stds_process_2 / sqrt(10))
maxs_process_3 = means_process_3 + (1.96 * stds_process_3 / sqrt(10))
maxs_replaying = means_replaying + (1.96 * stds_replaying / sqrt(10))
maxs_replaying_2 = means_replaying_2 + (1.96 * stds_replaying_2 / sqrt(10))
maxs_replaying_3 = means_replaying_3 + (1.96 * stds_replaying_3 / sqrt(10))
maxs_coruntine = means_coruntine + (1.96 * stds_coruntine / sqrt(10))
maxs_coruntine_2 = means_coruntine_2 + (1.96 * stds_coruntine_2 / sqrt(10))
maxs_coruntine_3 = means_coruntine_3 + (1.96 * stds_coruntine_3 / sqrt(10))
;
```

Make plot.

```julia
layer_errorbar = layer(
    ymax = [1, 3, 5] .+ 0.25, ymin = [1, 3, 5] .- 0.25,
    x = [means_coruntine, means_replaying, means_process],
    xmin = [mins_coruntine, mins_replaying, mins_process],
    xmax = [maxs_coruntine, maxs_replaying, maxs_process],
Geom.errorbar, Theme(default_color=colorant"lightsteelblue")
)

m_p = round(means_process + 0.001, 3) * 100
m_p_2 = round(means_process_2 + 0.001, 3) * 100
m_p_3 = round(means_process_3 + 0.001, 3) * 100
m_r = round(means_replaying + 0.001, 3) * 100 # 0.0001 is for correct presion showing in plot
m_r_2 = round(means_replaying_2 + 0.001, 3) * 100
m_r_3 = round(means_replaying_3, 3) * 100
m_c = round(means_coruntine, 4) * 100
m_c_2 = round(means_coruntine_2, 4) * 100
m_c_3 = round(means_coruntine_3, 4) * 100

layer_label = layer(
    y = [5.5, 5, 4.5, 3.5, 3, 2.5, 1.5, 1, 0.5],
    x = [means_process_3, means_process_2, means_process, means_replaying_3, means_replaying_2, means_replaying, means_coruntine_3, means_coruntine_2, means_coruntine],
label = ["$m_p_3%", "$m_p_2%", "$m_p%", "$m_r_3%", "$m_r_2%", "$m_r%", "$m_c_3%", "$m_c_2%", "$m_c%"], Geom.label(position=:right)
)

layer_process_1 = layer(
    ymax = [4.5 + 0.25], ymin = [4.5 - 0.25], x = [means_process], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"skyblue")
)

layer_process_2 = layer(
    ymax = [5 + 0.25], ymin = [5 - 0.25], x = [means_process_2], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"skyblue3")
)

layer_process_3 = layer(
    ymax = [5.5 + 0.25], ymin = [5.5 - 0.25], x = [means_process_3], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"skyblue4")
)

# Coroutine replaying
layer_replaying_1 = layer(
    ymax = [2.5 + 0.25], ymin = [2.5 - 0.25], x = [means_replaying], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"hotpink")
)

layer_replaying_2 = layer(
    ymax = [3 + 0.25], ymin = [3 - 0.25], x = [means_replaying_2], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"hotpink3")
)

layer_replaying_3 = layer(
    ymax = [3.5 + 0.25], ymin = [3.5 - 0.25], x = [means_replaying_3], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"hotpink4")
)

# Coroutine copying
layer_coroutine_1 = layer(
    ymax = [0.5 + 0.25], ymin = [0.5 - 0.25], x = [means_coruntine], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"greenyellow")
)

layer_coroutine_2 = layer(
    ymax = [1 + 0.25], ymin = [1 - 0.25], x = [means_coruntine_2], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"green3")
)

layer_coroutine_3 = layer(
    ymax = [1.5 + 0.25], ymin = [1.5 - 0.25], x = [means_coruntine], 
    Geom.bar(orientation=:horizontal),
    Theme(default_color=colorant"green4")
)

# Plot
p = plot(
    layer_label,
    layer_errorbar,
    layer_coroutine_1, layer_coroutine_2, layer_coroutine_3,
    layer_replaying_1, layer_replaying_2, layer_replaying_3, 
    layer_process_1, layer_process_2, layer_process_3,
    Stat.yticks(ticks=[0, 1, 2, 3, 4]),
    Stat.xticks(ticks=[0.00, 0.25, 0.50, 0.75, 1.00]),
Guide.title("Percentage of Times Spent on Program State Copying"),
Guide.manual_color_key(
"",["POSIX Forking   ","Coroutine Replaying   ", "Coroutine Copying"], 
["skyblue3", "hotpink3", "green3"]),
Theme(key_position=:top, key_label_font_size=9.5pt),
    Guide.yticks(label=false),
Guide.ylabel("Method of Copying"), 
    Guide.xlabel("Program State Copying / Total (%)"),

    Coord.Cartesian(xmax=1.00),
Scale.x_continuous(labels=x->"$(round(Int, x*100))%")
)
```
